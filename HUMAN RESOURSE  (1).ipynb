<h2 style='color:brown;'>7. Fill Missing Values</h2>

<p><b>Problem:</b><br>
The dataset contains missing values in important columns such as salary, department, city, 
or performance_rating, which may affect analysis accuracy and decision-making.</p>

<p><b>Solution:</b><br>
Handle missing values by filling numerical columns using mean or median, 
categorical columns using mode, and date columns using appropriate default values 
to ensure data completeness and reliability.</p>

```python
df['salary'] = df['salary'].fillna(df['salary'].median())
df['department'] = df['department'].fillna(df['department'].mode()[0])
df['termdate'] = df['termdate'].fillna('Active')
```

```python
df.fillna({
    'salary': df['salary'].median(),
    'department': df['department'].mode()[0],
    'termdate': 'Active'
}, inplace=True)
```

```python
df.isna().sum()
```

<h2 style='color:brown;'>8. Removing Duplicate Records</h2>

```python
print("Duplicate Records:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicate Records After Removal:", df.duplicated().sum())
```

<h2 style='color:brown;'>10. Data Standardizing</h2>

<h3 style='color:green;'>Text Standardization</h3>

<ul>
<li>Convert all text to a uniform case (lower/title case).</li>
<li>Remove extra spaces from text fields.</li>
<li>Correct inconsistent spellings in data.</li>
<li>Remove unwanted special characters.</li>
</ul>

<h4 style='color:green;'>1. Remove Extra Spaces</h4>

```python
df['department'] = df['department'].str.strip()
df['city'] = df['city'].str.strip()
df['job_title'] = df['job_title'].str.strip()
df['first_name'] = df['first_name'].str.strip()
df['last_name'] = df['last_name'].str.strip()
```

<h4 style='color:green;'>2. Convert Text to Lowercase</h4>

```python
df['department'] = df['department'].str.lower()
df['city'] = df['city'].str.lower()
df['job_title'] = df['job_title'].str.lower()
df['gender'] = df['gender'].str.lower()
```

<h4 style='color:green;'>3. Remove Special Characters</h4>

```python
import re

df['department'] = df['department'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df['city'] = df['city'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df['job_title'] = df['job_title'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

<h4 style='color:green;'>4. Remove Multiple Spaces</h4>

```python
df['department'] = df['department'].str.replace(r'\s+', ' ', regex=True)
df['city'] = df['city'].str.replace(r'\s+', ' ', regex=True)
df['job_title'] = df['job_title'].str.replace(r'\s+', ' ', regex=True)
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.replace(r'\s+', ' ', regex=True)
```

<h4 style='color:green;'>5. Create Full Name Column</h4>

```python
df['full_name'] = df['first_name'] + ' ' + df['last_name']
df['full_name'] = (
    df['first_name'].fillna('') + ' ' + df['last_name'].fillna('')
).str.strip()
df[['first_name', 'last_name', 'full_name']].head()
```

<h2 style="color:Blue;">Salary Distribution by Department (Plotly)</h2>

<p><b>Insight:</b>The salary distribution by department shows how employee compensation varies across different departments in the organization..</p>

<h2 style="color:Blue;">Salary Share by Department</h2>

```python
import plotly.express as px

# 1. Sum salary by department
salary_share = (
    df
    .groupby('department')['salary']
    .sum()
    .reset_index()
)

# 2. Create pie chart
fig = px.pie(
    salary_share,
    names='department',
    values='salary',
    title='Total Salary Distribution by Department'
)

fig.show()
```
